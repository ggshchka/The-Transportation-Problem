```python
import pandas as pd
import numpy as np
```

```python
demand = np.array([50,100,75,75, 100, 50, 100])
supply = np.array([100,200, 100, 150])
c = [] 
for i in range(len(supply)):
    row = []
    for j in range(len(demand)):
        row.append('a'+ str(i+1) + str(j+1))
    c.append(row)
    
costs = np.array(c) #cost illustration
```

```python
costs
```

```python
def illstr_table(dmnd, sply):
    i_table = pd.DataFrame(costs, columns=dmnd, index=sply)
    for i in range(len(dmnd)):
        i_table.rename(columns={'0': str(dmnd[i])})
    return i_table
```

```python
illstr_table(demand, supply) #illustration
```

```python
def check_balanced(dmnd, sply):
    return "Balanced" if sum(dmnd)==sum(sply) else "Unbalanced"
```

```python
check_balanced(demand, supply)
```

```python
display(illstr_table(demand, supply).replace(costs, '-'))
```

```python
def solution(dmnd, sply):  #есть лишние шаги в алгоритме
    fg = 0
    cap_coord = []
    table = illstr_table(dmnd, sply)
    display(table.replace(costs, '-'))
    for i in range(len(sply)):
        for j in range(len(dmnd)):
            if j != fg:
                continue
            fg += 1
            cell_v = min(sply[i], dmnd[j])
            sply[i] -= cell_v
            dmnd[j] -= cell_v
            table.index = sply
            table.columns = dmnd
            cap_coord.append([cell_v, [i,j]])
            if sply[i] == 0:
                fg = j
                if i != len(sply)-1:
                    table = table.replace(costs[i][j], cell_v).replace(costs[i, :], '+')
                else:
                    table = table.replace(costs[i][j], cell_v).replace(costs[i, :], '+')
                display(table.replace(costs, '-'))
                break
            else: 
                table = table.replace(costs[i][j], cell_v).replace(costs[:, j], '+')
                display(table.replace(costs, '-'))
                #table = table.replace(costs[i][j], cell_v).replace(costs[i, :], '+')
    return cap_coord
```

```python
d = np.copy(demand)
s = np.copy(supply)
```

```python
solution(d, s)
```

```python
gen_data = np.random.randint(1, 10, len(supply)*len(demand))
```

```python
my_costs = np.reshape(gen_data, (-1, len(demand)))
my_costs
```

```python
def get_ans(cap_coord, csts):
    summ=0
    for cap, coord in cap_coord:
        summ += csts[coord[0]][coord[1]]*cap
    return summ
```

```python
print("Cost: ", get_ans(solution(demand, supply), my_costs))
```
